## Additional cleanup

```python
# 4 Añadir uid a las filas
import hashlib  # Libreria dentro de libreria estandar, para operaciones criptograficas
# Numero unico para poder identificar a nuestra fila

uids = (el_universal
            .apply(lambda row: hashlib.md5(bytes(row['url'].encode())), axis=1)
            # Aplicamos la funcion a las filas (axis=1). La función md5 de la librería hashlib no debe ser utilizada para criptografia. Lo utilizamos porque nos puede dar un identificador unico. Nos da un numero de 128 bytes que lo vamos a representar en hexadecimal. Lo codificamos en utf-8 gracias a la función .encode(). 
            .apply(lambda hash_object: hash_object.hexdigest())
            # El hash object obtenido anteriormente lo conventimos en un numero hexadecimal 

)

el_universal['uid'] = uids  # Añadimos el hash generado al dataframe
el_universal.set_index('uid', inplace=True) # Ahora queremos que esta sea nuestro indice.

el_universal
```

```python
# Eliminamos los saltos de línea de nuestros ariculos contenidos en la columna body
stripped_body = (el_universal
                    .apply(lambda row: row['body'], axis=1)
                    # Aplicamos una modificacion a cada una de las filas. La forma de obtener estas filas es diciendo que el axis=1
                    .apply(lambda body: list(body))  # Seleccionamos la columna body y convertimos cada fila en una lista de letras.
                    .apply(lambda letters: list(map(lambda letters: letters.replace('\n', ''), letters))) 
                    # Por cada letra en la lista de una fila del body reemplazamos en esta los saltos de lineas por espacios.
                    # Esto lo convertimos en un map para que podamos pasarle la lista con la que trabajará.
                    # Convertimos el objeto map entregado por la funcion map en un objeto lista.
                    .apply(lambda letters: ''.join(letters))
                    # Unimos las letras de cada lista para que obtener finalmente un string por fila

                )
stripped_body
```

## Data enrichment

```python
# 6. Tokenizar el título y el body
import nltk

from nltk.corpus import stopwords  # Palabras que no nos sirven o aportan valor significativo a una oración 

# nltk.download('punkt') Libreria para tokenizar, es decir dividir en palabras.
# nltk.download('stopwords') Palabras que no nos sirven o aportan valor significativo a una oración 

stop_words = set(stopwords.words('spanish'))

def tokenize_columns(df, column_name):
    return (df
                .dropna()  # Eliminamos todas las celdas que tengan NaN
                .apply(lambda row: nltk.word_tokenize(row[column_name]), axis=1)
                # De la librería nltk aplicamos la función work_tokenize a cada fila, pasando como parámetro el nombre de la columna.
                .apply(lambda tokens: list(filter(lambda token: token.isalpha(), tokens)))
                # Eliminamos todas las palabras que no sean alfanuméricas. Generamos un filtro y le decimos que sólo queremos las palabras que sean alfanumerica con la funcion .isalpha() y le pasamos la lista. La función filter() devuelve un objeto, por lo tanto debemos convertirlo en un objeto lista.
                .apply(lambda tokens: list(map(lambda token: token.lower(), tokens)))
                # Convertimos todos los tokens a lower case para que se puedan comparar con posterioridad
                .apply(lambda word_list: list(filter(lambda word: word not in stop_words, word_list)))
                # Eliminamos las palabras stopwords. Filtraremos todas aquellas palabras que no sean stop words y le pasamos una lista con todas las palabras. La función filter() devuelve un objeto, por lo tanto debemos convertirlo en un objeto lista.
                .apply(lambda valid_word_list: len(valid_word_list))
                # No queremos una lista de palabras, lo que queremos es saber la cantidad de palabras validas. 
            )

el_universal['n_tokens_title'] = tokenize_columns(el_universal, 'title')
el_universal['n_tokens_body'] = tokenize_columns(el_universal, 'body')

el_universal
```

El **clarín** y **el pais** tienen una política contraria, donde se enfocan en artículos de mayor extensión. Tal vez con mayor contenido informativo.
