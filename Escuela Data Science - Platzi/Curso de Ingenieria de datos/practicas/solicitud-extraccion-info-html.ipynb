## Extracci√≥n de informaci√≥n del HTML

```python
# Convirtiendo a UTF-8
response.encoding ='utf-8'
```

```python
import bs4

# Generamos un parser, el cual nos permitira hacer queries al texto
soup = bs4.BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.title.text)
```

```python
# soup nos regresa lista de nodos
print(soup.select('meta [name=description]')[0]['content'])
```

```python
#courses_link = soup.select('.RecentCourses-item')
courses_link = soup.select('.HomeCategories-anchor')

courses = [course['href'] for course in courses_link]

for i in courses:
    print(i)
```
