## Tools for thinking about design (Alberto Cairo)
En el libro el Arte Funcional, Alberto Cairo proporciona una herramienta parapensar en las compensaciones de diseño a la hora de construir graficos informativos



Dentro del circulo hay dimensiones que describen compensaciones entre dos enfoques, siendo una dimension mas compleja y profunda y la otra mas superficial pero entendible. Esta es una herramienta para que los diseñadores piensen en sus visualizaciones pero no es una herramietna de analitica por sí misma. Sin embargo, pensar en ello permite que las necesidades de cualquier problema en particular puedan cambiar cuales de estas dimensiones son importantes o podrian introducir nuevas dimensiones para pensar

**¿ Cómo alguien de ciencia de datos, a quien intentas alcanzar con tus visualizaciones?**
 Vamos a profundizar en algunas visualizaciones de Cairo: 
 * **Abstraction - Figuration**: Boxes and charts (abstraction) or real-world physical objects (figuration). Una visualizacion altamente figurativa describe el fenomeno utilizando representaciones fisicas, como fotograficas o dibujos, a medida que las representaciones se vuelven menos reales y mas conceptuales el énfases cambia de la figuracion a la abstraccion
 * **Functionality - Decoration** Un grafico completamente funcional no tiene adornos y esta mas cerca de una representcion directa de los datos. Mientras que un grafico muy decorado tiene adornos mas artisticulos. No es una pregunta de qué está bien y qué esta mal, los adornos pueden aumentar la cantidad de tiempo que un espectador pasa considerando la representacion. Con esto, se orienta a explorar sus matices y formar asociaciones mentales que pueden aumentar la familiaridad y memorabilidad
 * **Density-Lightness**: Se relaciona con la cantidad de informacion que se muestra. Hay un monton de ejemplos de visualizaciones cientificas donde las figuras estan destinadas a ser estudiadas a profundidad mientras que otras a aumentar rapidamente una narrativa. 
 * **Multidimensiona-Unidimensional**: Un grafico multidimensional describe un fenomeno en su conjunto e invita al espectador a explorar muchos aspectos diferentes de los fenomenos. En su lugar, un grafico unidimenional se centra en uno o varios elementos y los explora de una o mas maneras.
 * **Originality - Familiarity**: En el mundo moderno estamos acostumbrados a ver una cantidad de diferentes tipos de grafico con informacion. Pensar en terminos de estas representaciones se enseña a una edad muy temprana, esto los hace familiares a una amplia poblacion. Pero esto no siempre fue asi, es seguro que el grafico de barras es bastante familiar para la mayoria de las personas pero un grafico con mas originalidades, elementos que necesitan ser estudiado explicados, por ejemplo, el siguiente grafico que describe la marcha de Napoleón hacia Rusia en 1812

* **Novelty-Redundancy** La redundancia es la tendencia de un grafico a contar la misma historia de muchas maneras. Por ejemplo, un grafico de barras pero donde los colores tambien indiquen las escalas.

## Graphical Heuristics: Data-ink Ratio (Edward Tufte)



Una heuristica es un proceso o regla que esta destinado a guiarlo en la toma de decisiones; es por definicion no conocido por ser optimo o perfecto, sino por ser practico en la naturaleza. La heuristica esta destinada a ser seguidas hasta que tengas una razon para desviarte de ellas. La primera heuristica es data-ink (tinta de los datos). Tufte la define como el nucleo no borrable de un grafico. La tinta no redundante dispuesta en respuesta a la variacion en los numeros representados, es esencial para el proceso de toma de sentido de una variable dada.

Se define como la cantidad de tinta de datos dividida por la tinta necesaria para imprimir el grafico. Esto no quiere sugerir que debamos cuantificar la cantidad de tinta para imprimir un grafico, sino para orientarnos a eliminar elementos que no agregan nueva informacion al grafico. ## Graphical Heuristics: Chart junk (Edward Tufte)

Edward Tufte es mucho mas condenatorio, el sugiere que las decoraciones artisticas en graficos estadisticos son como malas hierbas en nuestros graficos de datos. Sugirió que en realidad hay tres tpos de chartjunk:
* El arte optico no intencionado: Por ejemplo, el sombreado excesivo o el patron de caracteristicas del gráfico. Vemos el siguiente ejemplo que tiene tantos patrones que el ojo salta y se fatiga. Por eso es poco comun utilizar, por ejemplo, camisas a rayas en cursos virtuales. Este fenomeno se llama Patrones de Moiré


En estos casos seria mejor etiquetar directamente los graficos.
* La cuadrícula: Sugiere que la cuadricula es innecesaria pero tambien causa competencia con los datos reales que se comparten. El adelgazamiento, la eliminacion o la desaturacion de las lineas de cuadricula facilita la visualizacion de los datos, en lugar de verse abrumado por el numero de lineas de la pagina. El etiquetado directo de los datos es otra gran manera de reducir esta forma de charjunk
* El pato: Terminos generales se refieren a graficos creativos que no son datos, ya sean de arte de linea o fotografia que estan en el grafico. Los periodicos y las revistas de noticias son un lugar donde se utiliza a menudo este tipo de imagenes


Este grafico muestra la tendencia del precio de los diamantes entre 1978 y 1982. Aunque las cnatidades se olvidan facilmente, es facil de recordar que la tendencia tiene un pico debido a la forma de la pierna de la mujer (Estos casos se llaman Holmes charts)

## The truthfull Art

Libro que debe ser una lectura obligatoria para cualquier datascientist. Empaca referencias a expertos y detalles en lecciones memorables.
Habla de cualidades que ester interrelacionadas


* Veraz: Como cientificos de datos tenemos que ser personas que intenten mantener la verdad. Tenemos que ser honestos con nosotros mismos cuando limpiamos y resumimos datos porque es posible que nuestras actividades oculten un mensaje por las limitaicones que hemos aplicado. Debemos ser explicitos con cada cambio que le aplicamos a los datos y evitar caer en el autoengaño. A veces nos centramos tanto en la busqueda de patrones que reducimos y presentamos datos que no son autenticas para los fenomenos que describen.
    * Evitar engañarnos nosotros mismos
    * Evitar engañar a nuestra audiencia 
* Functionality: Hacer los gráficos funcionales que sean entendibles para la audiencia. Recordemos el ejemplo anterior de darkhorse donde cambiamos un gráfico a su version minimalista y poniendo los valores en las barras para que pudieran hacer comparaciones facilmente. Comprender cuan funcional es un enfoque determinado de una visualizacion no requiere necesariamente una investigacion costosa. Ahora es mucho mas facil probar hipotesis directamente con fuentes de multitutes
* Beauty: Aquí hay que conocer un poco mas a la audiencia. Hay diferentes percepciones basadas en las experiencias vividas de las personas. Hay muco para ocnstruir una hermosa experiencia de visualziacion de la informacion que tambien depende de factores situacionales y contextuales.  
* Insightful: Los graficos deben ser perspicaces. No deberian solo replicar datos de tablas, sino que deberian dibujar al espectador de tal manera que el espectador tenga un momento _eureka_. La grafica de informacion debe ser esclarecedora.
* Esclarecedora: Esta cualidad de compone de las 4 anteriores. An enlightening graphic is a consequence of paying attention to the previous four qualities. A graphic that is truthful, functional, beautiful, and insightful has the potential of being enlightening as well
